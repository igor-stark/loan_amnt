```python
path = 'dados/Case_Data.csv'
df = pd.read_csv(path)
```

```python
df_ = df.copy()
```

```python
target_cols = ['id', 'loan_amnt', 'sub_grade', 'title']
df_ = df_.loc[:, target_cols].reset_index(drop=True)
```

```python
rows = 20000
df_ = df_.iloc[:rows,:]
```

```python
seed = 42
scalar = 100000
np.random.seed(seed)
random_numbers = np.random.rand(len(df_['loan_amnt']))
df_['loan_amnt'] = df_['loan_amnt'] * scalar * random_numbers
```

```python
df_ = df_.sort_values(by='loan_amnt', ascending=False)
```

```python
ratings = df_['sub_grade'].sort_values(ascending=True).unique()[:25]
smaller_df = df_[df_['sub_grade'].isin(ratings)].reset_index(drop=True)
```
